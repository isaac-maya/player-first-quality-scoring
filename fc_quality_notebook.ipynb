# FC Quality Notebook

Synthetic player-experience quality scoring across the five categories in `defect_taxonomy.md`. The same prioritization rule used in `fc_quality_scorer.py` is shown here in a compact analytics view.

Rule of thumb: trust impact and breadth lead, frequency follows, recovery cost adjusts.

```python
issues = [
    {"id": "FC-101", "category": "Input responsiveness", "frequency": 42, "trust": 5, "repro": 4, "recovery": 2},
    {"id": "FC-102", "category": "Progression/reward trust", "frequency": 11, "trust": 5, "repro": 5, "recovery": 1},
    {"id": "FC-103", "category": "UI flow clarity", "frequency": 65, "trust": 2, "repro": 5, "recovery": 5},
    {"id": "FC-104", "category": "Matchmaking fairness", "frequency": 24, "trust": 4, "repro": 3, "recovery": 3},
    {"id": "FC-105", "category": "Stability/performance", "frequency": 8, "trust": 5, "repro": 2, "recovery": 1},
]

for issue in issues:
    issue["priority_score"] = (
        issue["trust"] * 12
        + issue["frequency"] * 0.4
        + issue["repro"] * 4
        - issue["recovery"] * 2
    )

ranked = sorted(issues, key=lambda item: item["priority_score"], reverse=True)
for item in ranked:
    print(f"{item['priority_score']:6.1f}  {item['id']}  {item['category']:<26}  trust={item['trust']}  freq={item['frequency']}")
ranked
```

## Interpretation

The scoring intentionally weights player trust heavily. A reward-trust or stability issue can outrank a more common UI issue because progression credibility and session continuity are harder to recover once damaged.
